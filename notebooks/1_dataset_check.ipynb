## –ó–∞–≥—Ä—É–∑–∫–∞ –∏ –ø–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –¥–∞–Ω–Ω—ã—Ö

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
db = pd.read_parquet('chgk_questions.parquet')
gq = pd.read_parquet('got_q_db.parquet')
```

```python
db = db[db['question_type'] == '–ß']
```

```python
db
```

```python
gq
```

```python
db['package_played_at'] = pd.to_datetime(db['package_played_at'])
gq['date'] = pd.to_datetime(gq['date'])
```

## –°—Ä–∞–≤–Ω–µ–Ω–∏–µ –≥–æ–¥–æ–≤–æ–π –¥–∏–Ω–∞–º–∏–∫–∏

```python
sns.set_theme(style="whitegrid", palette="viridis")
fig, ax = plt.subplots(figsize=(12, 7)) 

yearly_counts_db = db.set_index('package_played_at')['question_id'].resample('YE').count()
yearly_counts_db.plot(
    ax=ax,              
    marker='o',          
    linestyle='-',             
    markersize=8,
    label='–î–∞–Ω–Ω—ã–µ –∏–∑ db'  
)

yearly_counts_gq = gq.set_index('date')['id'].resample('YE').count()
yearly_counts_gq.plot(
    ax=ax,               
    marker='s',          
    linestyle='--',      
    markersize=7,
    label='–î–∞–Ω–Ω—ã–µ –∏–∑ gq' 
)

plt.title('–°—Ä–∞–≤–Ω–µ–Ω–∏–µ –≥–æ–¥–æ–≤–æ–π –¥–∏–Ω–∞–º–∏–∫–∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ —Å—ã–≥—Ä–∞–Ω–Ω—ã—Ö –≤–æ–ø—Ä–æ—Å–æ–≤', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('–ì–æ–¥', fontsize=12)
plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤–æ–ø—Ä–æ—Å–æ–≤', fontsize=12)
plt.xticks(rotation=0) 
plt.legend() 

plt.tight_layout()
plt.show() 
```
